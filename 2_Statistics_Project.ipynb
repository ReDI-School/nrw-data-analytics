# Project - an historical Statistical Test

--- 
## The description:
Welcome to second first project! Have you heard about Ronald Aymer Fisher? He is the founder of the modern frequentist statistics and experiment randomization. This is the type of statistics that is still used today e.g for the market authorization of drugs and vaccines. Although Fisher roots his work in very analytical methods, he is probably also one of the most controversial figure of science and polymath in the 20th century from today's point of view, especially when it comes to his work as a geneticist.

## The Task:
1. How many cups of tea with milk does one have give to a tea connoisseur to blind-taste in order to make sure that this person can truly distinguish those cups in which milk has been poured after the tea from those for which the order has been opposite?
 
Answer the question with (possibly recent) publications and results.
