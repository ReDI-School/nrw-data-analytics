## The description:
Have you heard about Immobilienscout 24?  [Immobilienscout 24](https://https://www.immobilienscout24.de/). It the biggest platform for renting appartments in Germany. The company hires data analysts and data scientists for analyzing the market data. They are especially interested in the Berlin market. The company wants to understand where to publish advertisment.

--- 
## The dataset:
The datasets contains information about rental appartments in Berlin. 
One row is for one appartment. The dataset contains 6 columns.  
- Region: is the Berlin district, where the flat is located
- Condition: is the condition of the flat
- Rooms: The number of rooms the flat has
- Rent: Monthly rent for the flat
- Year_Construction: The year in which the house was build
- Space: How many square meters does the flat has?

```
import pandas as pd
```

```
immo = pd.read_csv("https://raw.githubusercontent.com/juliandnl/redi_ss20/master/berlin_rental.csv")
print(immo.sample(5))
```

```
immo.info()
```

# 1.How many apartments are there in Mitte in a 'well-kept' condition?

```
well_kept_mitte = immo[(immo['Region'] == 'Mitte') & (immo['Condition'] == 'well_kept')]
len(well_kept_mitte)
```

```
immo[(immo['Region'] == 'Mitte') & (immo['Condition'] == 'well_kept')]
```

# 2.How many apartments are there in Kreuzberg built before 1900?

```
kreuzberg_before_1900 = immo[(immo['Region'] == 'Kreuzberg') & (immo['Year_Construction'] < 1900)]
kreuzberg_before_1900
```

# 3. What are the average spaces of the apartments having 4 rooms?

```
immo[(immo['Rooms'] == 4)].groupby('Rooms')[['Space']].mean().round()
```

```
immo.groupby('Rooms')[['Space']].mean().round()
```

# 3.What is the average space of the apartments & number of rooms per region?

```
immo.groupby('Region')[['Space','Rooms']].mean().round()
```

```
#  FutureWarning: Indexing with multiple keys (implicitly converted to a tuple of keys) will be deprecated, use a list instead.
print(immo.groupby('Region')['Space', 'Rooms'].mean().round(1).sort_values('Space',ascending= False))
```

```
print(immo.groupby('Region')['Space', 'Rooms'].mean().round(1).sort_values('Space',ascending= False))
```

# 4.Which is the most expensive region on average and what is the maximum rent for this region?

```
# 1) Step-wise
# most expensive region on average: Mitte
mean_rent = immo.groupby('Region')[['Rent']].mean().round(2).sort_values('Rent', ascending=False)
print(mean_rent)

# maximum rent for this region:
rent_mitte = immo[(immo['Region'] == 'Mitte')]
rent_mitte.max()
```

```
# 2) Aggregated
# together:
mean_rent_max = immo.groupby('Region')[['Rent']].agg(['mean','max'])
mean_rent_max

# renamed multi-indexed columns and sorted by max_rent
mean_rent_max_2 = immo.groupby('Region')[['Rent']].agg([('mean_rent','mean'),('max_rent','max')]).sort_values([('Rent', 'max_rent')], ascending=False)
mean_rent_max_2
```

# Bonus Question:
5. What are the min, max and average number of rooms of the apartments in Köpenick?


```
rooms_koepenick = immo.groupby('Region')[['Rooms']].agg(['mean', 'min', 'max'])
rooms_koepenick
```

#  6. What is the average price per square meter for different Regions?

```
# a) new column: price_sq_meter (calculated as rent/ space)
def calculate_price_sq_meter(row):
    calc = row.Rent / row.Space
    return calc

immo['price_sq_meter'] = immo.apply(calculate_price_sq_meter, axis=1)
immo['price_sq_meter'] = immo['price_sq_meter'].round(2)
immo
```

```
# b) get mean price_sq_meter & group by region
immo.groupby('Region')['price_sq_meter'].mean().round(2).sort_values(ascending=False)

# Mitte has the highest price with 25.97€ per square meter
```

# 7. Is there a difference of average room space (`Space` divided by `Rooms`) for apartments built after 2000 versus apartment built before 2000?

```
# a) calculate the average room space 
def avg_room_space (row):
    return row.Space / row.Rooms

immo['avg_room_space'] = immo.apply(avg_room_space, axis=1)
immo
```

```
# b) Problem: to many years to display
immo.groupby('Year_Construction')['avg_room_space'].mean()
```

```
# b) group by  apartments built before 2000 and after
apartments_before_2000  = immo.query('Year_Construction < 2000')
print('The average room space for apartments built before 2000 is ' + str(apartments_before_2000.avg_room_space.mean()))

apartments_in_after_2000  = immo.query('Year_Construction >= 2000')
print('The average room space for apartments built in or after 2000 is ' + str(apartments_in_after_2000.avg_room_space.mean()))

# -> newer apartments have smaller rooms
```

# 8. What's the percentage of the apartments in high quality condition? 

In order to answer this question let's assume that high quality conditions are: `first_time_use`, `mint_condition`, `refurbished`, `first_time_use_after_refurbishment` and `fully_renovated`. The rest we can categorize as low quality condition.

```
# 1) create column with value high_quality when condition is ...., else low_quality

def high_low_quality (row):
  if row['Condition'] in ['first_time_use', 'mint_condition', 'refurbished', 'first_time_use_after_refurbishment', 'fully_renovated']:
    val = 'high_quality' 
  else:
    val = 'low_quality'
  return val

immo['quality'] = immo.apply(high_low_quality, axis=1)
```

```
# 2) calculate the percentage
len(immo[(immo['quality'] == 'high_quality')])/ len(immo)*100
```

```
# 3) check distribution high versus low quality
immo['quality'].value_counts()
```
